```python
import torch

from LoraCollisionLSTM import LoraCollisionLSTM 
from LoraCollisionTransformer import LoraCollisionTransformer

model_path = '/home/ylnner/Documents/new_model/scripts/results_server_0506_bilstm/scripts/models/bilstm_best_seed1.pth'
# /home/ylnner/Documents/new_model/scripts/results_server_0506_bilstm/scripts/models/bilstm_best_seed1.pth

# BiLSTM best config
#best_config = {'hidden_size': 64, 'num_layers': 3, 'dropout': 0.1, 'batch_size': 32}     
# {'hidden_size': 32, 'num_layers': 3, 'dropout': 0.6, 'batch_size': 32}  
best_config = {'hidden_size': 32, 'num_layers': 3, 'dropout': 0.6, 'batch_size': 32}
model = LoraCollisionLSTM(hidden_size=best_config['hidden_size'], num_layers=best_config['num_layers'], dropout=best_config['dropout'])
weights = torch.load(model_path, map_location=torch.device('cpu'))
model.load_state_dict(weights)
model.eval()
scripted_model = torch.jit.script(model)
scripted_model.save("bilstm_scripted.pt")
print("Conversion complete!")



# Transfromer best config
#model_path = '/home/ylnner/Documents/new_model/scripts/results_server_220517/scripts/models/transformer_best_seed6.pth'
#model_path = '/home/ylnner/Documents/new_model/scripts/results_server_0506_v2/scripts/models/transformer_best_seed1_epoch7.pth'
model_path = '/home/ylnner/Documents/new_model/scripts/results_server_0506_v2/scripts/models/transformer_best_seed1.pth'

#best_config = {'d_model': 32, 'n_heads': 4, 'n_layers': 3, 'dropout': 0.1, 'lr': 0.001, 'batch_size': 32}
best_config = {'d_model': 32, 'n_heads': 4, 'n_layers': 1, 'dropout': 0.5, 'lr': 0.001, 'batch_size': 32}
model_transformer = LoraCollisionTransformer(num_numerical_features=8, d_model=best_config['d_model'], n_heads=best_config['n_heads'], n_layers=best_config['n_layers'], dropout=best_config['dropout'])

weights = torch.load(model_path, map_location=torch.device('cpu'))
model_transformer.load_state_dict(weights)

model_transformer.eval()
scripted_model = torch.jit.script(model_transformer)
scripted_model.save("transformer_scripted.pt")
print("Conversion complete!")
```
