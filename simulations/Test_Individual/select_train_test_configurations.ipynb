```python
for key in all_files.keys():
    print(f"Configuration {key}: {len(all_files[key])} files")
    files = all_files[key]
    num_files_train = int(len(files)*0.8)
    num_files_test  = len(files) - num_files_train
    print(f"Number of files for training: {num_files_train}")
    print(f"Number of files for testing: {num_files_test}")
        
    train_files_selected = random.sample(files, num_files_train)
    test_files_selected = [f for f in files if f not in train_files_selected]

    print(f"Selected training files for configuration {key}: {train_files_selected}")
    print(f"Selected testing files for configuration {key}: {test_files_selected}")
    print("\n")

```
